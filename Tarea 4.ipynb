```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data_casen = pd.read_csv("datos_jorge.csv")
```

¿Cuál fue el nivel educacional más alto alcanzado o el nivel educacional actual?

```python
conteos = data_casen['e6a'].value_counts()

desfase = [0] * len(conteos)
desfase[-1] = 0.1

plt.figure(figsize=(4, 4))
plt.pie(conteos.tolist(), labels=conteos.index.tolist(), autopct='%0.1f%%', explode=desfase)
plt.title(' ¿Cuál fue el nivel educacional más alto alcanzado o el nivel educacional actual?', y=1.1)
plt.axis('equal')
plt.show()

```

-La mayor cantidad de encuestados terminó la educacion media cientifico humanista o la está cursando.
-La educacion basica es el nivel maximo nivel de casi 1/4 de los encuestados.

¿Cuál es el tipo de pareja segun sexo en Chile? (Legal o de hecho)

```python
plt.figure(figsize=(4,4))
count_plot = sns.countplot(x = data_casen.sexo, hue = data_casen.pareja)
plt.title('Gráfico de barras para tipo de pareja segun sexo',y=1.01)
plt.ylabel('Conteo')
plt.xlabel('sexo')
plt.show()
```

-Mas del 50% de los encuestados no tiene pareja formal
-Las parejas legales son superiores a las de hecho

```python
¿Cuál es la densidad de las edades en chile?
```

```python

plt.figure(figsize=(4,4))
sns.distplot(data_casen.edad, hist=True, kde=True, #kde estimación de la densidad
             bins=8, color = 'blue')
plt.title('Histograma de la variable Edad',y=1.01)
plt.ylabel('Densidad')
plt.xlabel('Edad')
plt.show()
```

-La mayor densidad poblacional se encuentra entre los 15 a 25 años aproximadamente
-A partir de los 70 años la densidad poblacional disminuye drasticamente.

¿Cual es la cantidad de personas que viven en sectores rurales y urbanos?

```python
plt.figure(figsize=(4,4))
count_plot = sns.countplot(x = data_casen.region, hue = data_casen.zona)
plt.title('Gráfico de barras para distribucion zonal en cada region',y=1.01)
plt.ylabel('Conteo')
plt.xlabel('Region')
plt.show()
```

    -La poblacion urbana cuatriplica a la poblacion rural
    -La sexta region (o´higgins) es la con mayor presencia de sector rural
